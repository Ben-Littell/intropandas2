```python
import pandas as pd
import numpy as np

path = '../input/portland-weather/PortlandWeather2013b.txt'

wx_df = pd.read_csv(path, delim_whitespace=True) # delim_whitespace changes seperator from , to whitespace for a txt file
wx_df.columns = wx_df.columns.str.lower() # makes the column names lower
wx_df.drop(['station'], axis=1, inplace=True)
wx_df.drop([0], axis=0, inplace=True)
new_indices = pd.Series(np.arange(0, len(wx_df)))
wx_df.set_index(new_indices, inplace=True)
wx_df[['prcp', 'snwd', 'snow', 'tmax', 'tmin']] = wx_df[['prcp', 'snwd', 'snow', 'tmax', 'tmin']].astype('int32')
wx_df[['snwd', 'prcp', 'snow']] = wx_df[['snwd', 'prcp', 'snow']]/25.4
wx_df[['tmax', 'tmin']] = wx_df[['tmax', 'tmin']] / 10 * (9/5) + 32
wx_df['snwd'][364] = wx_df['snwd'][363]
wx_df['snow'][364] = 0
wx_df['avg_temp'] = (wx_df['tmax'] + wx_df['tmin'])/2
wx_df['date'] = pd.to_datetime(wx_df['date'])
wx_df['year'] = wx_df['date'].dt.year
wx_df['month'] = wx_df['date'].dt.month
wx_df['day'] = wx_df['date'].dt.day



wx_df
```
